# Data Validation

```python
# Taking care of null values
vgsales['Publisher'] = vgsales['Publisher'].fillna('Unknown')

# Updated data types
print(vgsales.shape)
vgsales.info()
```

```python
#Years after 2015 were removed as the data after that was so sparse that it did not help the study, although a new console came out that year
vgsales = vgsales[vgsales['Year'] <= 2015]

# Convert the year to integers, or at least make sure they are integers
vgsales = vgsales[vgsales['Year'].notna()]
vgsales['Year'] = pd.to_numeric(vgsales['Year'], errors='coerce').astype('int64')


# Removing nulls
vgsales = vgsales.dropna(how='all')
print(vgsales.shape)
vgsales.info()
```

```python
# Making sure there are no video games that do not correspond to your console and year 

# Make sure data is sorted by 'Year' to get the first game per console
vgsales_sorted = vgsales.sort_values(by='Year').dropna(subset=['Year'])

# The first game of each console is selected
first_game_per_platform = vgsales_sorted.groupby('Platform').first().reset_index()

# The desired columns were filtered and the result is displayed
first_game_per_platform = first_game_per_platform[['Platform', 'Name', 'Year']]
print(first_game_per_platform)
```

```python
# A game was identified that was not in the correct year

# Find the row where the game name is "Strongest Tokyo University Shogi DS" and update the year
vgsales.loc[vgsales['Name'] == 'Strongest Tokyo University Shogi DS', 'Year'] = 2007

# Making sure the data is sorted by 'Year' to get the first game per console
vgsales_sorted = vgsales.sort_values(by='Year').dropna(subset=['Year'])

# Again, the first game of each console is selected
first_game_per_platform = vgsales_sorted.groupby('Platform').first().reset_index()

# The desired columns are filtered and the result is displayed
first_game_per_platform = first_game_per_platform[['Platform', 'Name', 'Year']]
print(first_game_per_platform)
```

## Best-selling games by year.

```python
# Group by year, game name, and console to get sales by region
sales_by_game_region = vgsales.groupby(['Year', 'Name', 'Platform'])[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']].sum()

# Sum sales by game and year (adding sales across consoles)
sales_by_game_year = sales_by_game_region.groupby(['Year', 'Name']).sum()

# Now we sum global sales (total across all regions)
sales_by_game_year['Global_Sales'] = sales_by_game_year[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']].sum(axis=1)

# Get the top selling game for each year (based on global sales)
top_game_per_year = sales_by_game_year.loc[sales_by_game_year.groupby('Year')['Global_Sales'].idxmax()]

# Show the final table
top_game_per_year = top_game_per_year.reset_index()

# Show the table
top_game_per_year
```

## Platforms with the most revenue per year

```python
# Step 1: Sum sales by console and year for all regions and Global_Sales
sales_by_platform_year = vgsales.groupby(['Year', 'Platform']).agg({
    'NA_Sales': 'sum',
    'EU_Sales': 'sum',
    'JP_Sales': 'sum',
    'Other_Sales': 'sum',
    'Global_Sales': 'sum'
}).reset_index()

# # Step 2: Identify the top console in each region by year
def top_platform_by_region(region):
    return sales_by_platform_year.loc[sales_by_platform_year.groupby('Year')[region].idxmax()][['Year', 'Platform', region]].rename(columns={
        'Platform': region.split('_')[0],
        region: f"{region}_Sales"
    })

na_top = top_platform_by_region('NA_Sales')
eu_top = top_platform_by_region('EU_Sales')
jp_top = top_platform_by_region('JP_Sales')
other_top = top_platform_by_region('Other_Sales')
global_top = top_platform_by_region('Global_Sales')

# Step 3: Merge the results into a single table
top_platforms_table = (na_top
                       .merge(eu_top, on='Year')
                       .merge(jp_top, on='Year')
                       .merge(other_top, on='Year')
                       .merge(global_top, on='Year'))

# Display the table
top_platforms_table
```

### Chart 2: Sales Trends by Video Game Genre in Each Region

```python
# Define regions
regions = ['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']

# Ensure the year is numeric
vgsales['Year'] = vgsales['Year'].astype(int)

# Create a stacked DataFrame for regions and genres
stacked_data = vgsales.melt(
    id_vars=['Year', 'Genre'],
    value_vars=regions,
    var_name='Region',
    value_name='Sales'
)

# Group and pivot data to get the desired format
stacked_data = (
    stacked_data.groupby(['Year', 'Region', 'Genre'])['Sales']
    .sum()
    .reset_index()
    .pivot(index=['Year', 'Region'], columns='Genre', values='Sales')
    .fillna(0)
    .reset_index()
)

# Get the list of genres
genres = stacked_data.columns[2:]  # Exclude 'Year' and 'Region' from columns

# Create a dictionary to associate each genre with its color
genre_colors = {
    "Action": "#00AEFF",       # Light blue
    "Adventure": "#1C6AB6",    # Dark blue
    "Fighting": "#70B531",     # Light green
    "Misc": "#226C1E",         # Dark green
    "Platform": "#F96464",     # Light red
    "Puzzle": "#E31A1C",       # Dark red
    "Racing": "#FFB732",       # Light orange
    "Role-Playing": "#E87400", # Dark orange
    "Shooter": "#B18DC2",      # Light purple
    "Simulation": "#6A3D9A",   # Dark purple
    "Sports": "#7C3F00",       # Light brown
    "Strategy": "#3E1C00"      # Dark brown
}

# Chart configuration
fig, axes = plt.subplots(2, 2, figsize=(16, 10), sharex=True)
axes = axes.flatten()

for i, region in enumerate(regions):
    ax = axes[i]
    region_data = stacked_data[stacked_data['Region'] == region]
    bottom = None
    for genre in genres:
        color = genre_colors.get(genre, '#000000')  # Assign color based on genre, use black as a fallback
        ax.bar(
            region_data['Year'],
            region_data[genre],
            bottom=bottom,
            label=genre if i == 0 else None,  # Avoid duplicating labels in the legend
            color=color
        )
        bottom = region_data[genre] if bottom is None else bottom + region_data[genre]
    ax.set_title(region)
    ax.set_xlabel("Year")
    ax.set_ylabel("Sales (in millions)")
    ax.set_xticks(sorted(region_data['Year'].unique()))
    ax.set_xticklabels(sorted(region_data['Year'].unique()), rotation=90)

# Create a custom legend
legend_handles = [plt.Rectangle((0, 0), 1, 1, color=genre_colors[genre]) for genre in genres]
plt.legend(legend_handles, genres, title="Genre", bbox_to_anchor=(1.05, 0.15), loc='lower left', fontsize=9)
plt.suptitle("Sales Trend by Video Game Genre in Each Region")
plt.tight_layout(rect=[0, 0, 0.85, 1])
plt.show()
```

### Chart 3: Top 5 best-selling video games by genre

```python
# Create a dictionary to associate each genre with its color
genre_colors = {
    "Action": "#00AEFF",       # Light blue
    "Adventure": "#1C6AB6",    # Dark blue
    "Fighting": "#70B531",     # Light green
    "Misc": "#226C1E",         # Dark green
    "Platform": "#F96464",     # Light red
    "Puzzle": "#E31A1C",       # Dark red
    "Racing": "#FFB732",       # Light orange
    "Role-Playing": "#E87400", # Dark orange
    "Shooter": "#B18DC2",      # Light purple
    "Simulation": "#6A3D9A",   # Dark purple
    "Sports": "#7C3F00",       # Light brown
    "Strategy": "#3E1C00"      # Dark brown
}

# Group by genre and sort by global sales to get the top 5 best-selling games in each genre
sns.set_style('darkgrid', {"grid.color": ".80"})
top_5_games_by_genre = (
    vgsales.groupby('Genre')
    .apply(lambda x: x.nlargest(5, 'Global_Sales', keep='all')[['Genre', 'Global_Sales', 'Name', 'Platform']])
    .reset_index(drop=True)
)

# Add a column with the formatted name + platform
top_5_games_by_genre['Game_Label'] = top_5_games_by_genre['Name'] + ' (' + top_5_games_by_genre['Platform'] + ')'

# Set the chart size
plt.figure(figsize=(14, 14))

# Create a stacked bar chart
bar_height = 0.6
unique_genres = top_5_games_by_genre['Genre'].unique()

for genre in unique_genres:
    genre_data = top_5_games_by_genre[top_5_games_by_genre['Genre'] == genre]
    plt.barh(
        genre_data['Game_Label'], genre_data['Global_Sales'],
        label=genre,
        height=bar_height,
        color=genre_colors.get(genre, '#000000')  # Assign color based on genre, use black as a fallback
    )

# Labels and title
plt.xlabel('Global Sales (Millions)')
plt.ylabel('Video Games (Console)')
plt.title('Top 5 Best-Selling Video Games by Genre')
plt.legend(title='Genre', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

# Growth in the last 4 years

```python
# Filtered from the last 4 years

last_four_years = vgsales[(vgsales['Year'] >= 2012) & (vgsales['Year'] <= 2015)]
last_four_years.head(10)
```

## Gender with the highest and lowest growth in the last 4 years

```python
# Split into two periods: 2012-2013 and 2014-2015
first_period = last_four_years[last_four_years['Year'] < 2014]
second_period = last_four_years[last_four_years['Year'] >= 2014]

# Group and sum global sales by genre in both periods
genre_sales_1st_period = first_period.groupby('Genre')['Global_Sales'].sum()
genre_sales_2nd_period = second_period.groupby('Genre')['Global_Sales'].sum()

# Combine results and calculate percentage growth
genre_growth = pd.DataFrame({
    'First_Period_Sales': genre_sales_1st_period,
    'Second_Period_Sales': genre_sales_2nd_period
}).fillna(0)

genre_growth['Growth_Percent'] = (
    (genre_growth['Second_Period_Sales'] - genre_growth['First_Period_Sales'])
    / genre_growth['First_Period_Sales'].replace(0, 1)) * 100

# Identify the genre with the highest and lowest growth
max_genre_growth = genre_growth['Growth_Percent'].idxmax()
min_genre_growth = genre_growth['Growth_Percent'].idxmin()

# Print results:
print("Growth by Genre:")
print(f"Highest Growth: {max_genre_growth} - Growth: {genre_growth.loc[max_genre_growth, 'Growth_Percent']:.2f}%")
print(f"Lowest Growth: {min_genre_growth} - Growth: {genre_growth.loc[min_genre_growth, 'Growth_Percent']:.2f}%")
```

## Platform that has had the highest and lowest growth in the last 4 years

```python
# Split into two periods: 2012-2013 and 2014-2015
first_period = last_four_years[last_four_years['Year'] < 2014]
second_period = last_four_years[last_four_years['Year'] >= 2014]

# Group and sum global sales by platform in both periods
platform_sales_1st_period = first_period.groupby('Platform')['Global_Sales'].sum()
platform_sales_2nd_period = second_period.groupby('Platform')['Global_Sales'].sum()

# Combine results and calculate percentage growth
platform_growth = pd.DataFrame({
    'First_Period_Sales': platform_sales_1st_period,
    'Second_Period_Sales': platform_sales_2nd_period
}).fillna(0)

# Calculate percentage growth (avoiding division by zero)
platform_growth['Growth_Percent'] = (
    (platform_growth['Second_Period_Sales'] - platform_growth['First_Period_Sales'])
    / platform_growth['First_Period_Sales'].replace(0, np.nan)) * 100

# Identify the platform with the highest and lowest growth
max_platform_growth = platform_growth['Growth_Percent'].idxmax()
min_platform_growth = platform_growth['Growth_Percent'].idxmin()

# Print results
print("Growth by Platform:")
print(f"Highest Growth: {max_platform_growth} - Growth: {platform_growth.loc[max_platform_growth, 'Growth_Percent']:.2f}%")
print(f"Lowest Growth: {min_platform_growth} - Growth: {platform_growth.loc[min_platform_growth, 'Growth_Percent']:.2f}%")

```
